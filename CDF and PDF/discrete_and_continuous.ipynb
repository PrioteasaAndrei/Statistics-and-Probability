### Discrete random variables

```python
## function that "randomly" generates data
## we want to inspect the distribution of the output

def data_generator():
    initial_rand = random.randint(-1000,1000)

    cum_sum = initial_rand * 256 + 774
    if random.randint(1,10) < 4:
        cum_sum = cum_sum % 1000

    if cum_sum % 2 == 0:
        cum_sum += 231

    return cum_sum * random.randint(1,5) - random.randint(1,1000)

sample_size = 100000
sample = [data_generator() for i in range(sample_size)]

# Counting the occurrences of each value in the random variable
value_counts = {value: sample.count(value) for value in set(sample)}

# Calculating the probabilities for each value
total_outcomes = len(sample)
probabilities = {value: count / total_outcomes for value, count in value_counts.items()}
```

```python
# probabilities.values()

plt.scatter(list(probabilities.keys()),list(probabilities.values()))
```

```python
# Step 1: Visual Inspection (Histogram)
plt.hist(list(probabilities.keys()), bins=10, alpha=0.6, color='b', edgecolor='black')
plt.xlabel('Samples')
plt.ylabel('Frequency')
plt.title('Histogram of Probabilities')
plt.grid(True)
plt.show()
```
